<u> III Population dynamics: predator prey. </u>
1. Implement the following population dynamics model: 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import sympy as sp
import scipy.optimize as sc_opt
```

```python
def func(y, t, r, k, a, b, c, d):
    x, y = y
    dx_dt = r * x * (1 - x / k) - (a * x * y) / (c + x)
    dy_dt = (b * a * x * y) / (c + x) - d * y
    return [dx_dt, dy_dt]

a = 3.2 
b = 0.6
c = 50.0
d = 0.56
k = 125.0
r = 1.6

t = np.linspace(0, 50, 1000)
init = [1, 1] 
sol = odeint(func, init, t, args=(r, k, a, b, c, d))

plt.plot(t, sol[:, 0], label='x(t)')
plt.plot(t, sol[:, 1], label='y(t)')
plt.legend()
plt.xlabel('t')
plt.ylabel('x(t), y(t)')
plt.show()
```
