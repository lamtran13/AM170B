## County Time Series
> San Diego: Average Temperature, Tree Coverage Loss, Wildfire <br>
> Riverside: Wildfire and Ozone <br>
> Sonoma: Average Precipitation and Flood 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('California_official.csv')
display(data)
```

```python
# Sonoma Average Temperature
plt.rcParams['figure.figsize'] = (15, 10)
Sonoma = data.loc[(data['County'] == 'Sonoma')]
x = Sonoma['Date']
y = Sonoma['Avg Temp']
sns.lineplot(data=Sonoma, x='Date', y='Avg Temp', marker='o', mec='black',mfc='red')
for x, y in zip(Sonoma['Date'], Sonoma['Avg Temp']):
    plt.text(x = x, y = y, s = '{:.0f}'.format(y))
plt.xticks(data['Date'],rotation=45, fontsize=15)
plt.show()
```

```python
## San Diego Tree Coverage Loss
SD = data.loc[(data['County'] == 'San Diego')]
sns.lineplot(data=SD, x='Date', y='Tree Coverage Loss', marker='o', mec='black',mfc='red')
x = SD['Date']
y = SD['Tree Coverage Loss']
for x, y in zip(SD['Date'], SD['Tree Coverage Loss']):
    plt.text(x = x, y = y, s = '{:.0f}'.format(y))
plt.xticks(SD['Date'],rotation=45, fontsize=20)
plt.show()
```

```python
#San Diego Average Temperature
sns.lineplot(data=SD, x='Date', y='Avg Temp', marker='o', mec='black',mfc='red')
x = SD['Date']
y = SD['Avg Temp']
for x, y in zip(SD['Date'], SD['Avg Temp']):
    plt.text(x = x, y = y, s = '{:.0f}'.format(y))
plt.xticks(SD['Date'],rotation=45, fontsize=20)
plt.show()
```

```python
#San Diego Wildfire
SD = data.loc[(data['County'] == 'San Diego')]
sns.lineplot(data=SD, x='Date', y='Wildfire', marker='o', mec='black',mfc='red')
x = SD['Date']
y = SD['Wildfire']
for x, y in zip(SD['Date'], SD['Wildfire']):
    plt.text(x = x, y = y, s = '{:.0f}'.format(y))
plt.xticks(SD['Date'],rotation=45, fontsize=20)
plt.show()
```

```python
# Riverside Wildfire
RV = data.loc[(data['County'] == 'Riverside')]
sns.lineplot(data=RV, x='Date', y='Wildfire', marker='o', mec='black',mfc='red')
x = RV['Date']
y = RV['Wildfire']
for x, y in zip(RV['Date'], RV['Wildfire']):
    plt.text(x = x, y = y, s = '{:.3f}'.format(y))
plt.xticks(RV['Date'],rotation=45, fontsize=15)
plt.show()
```

```python
#Riverside Ozone
sns.lineplot(data=RV, x='Date', y='Ozone', marker='o', mec='black',mfc='red')
x = RV['Date']
y = RV['Ozone']
for x, y in zip(RV['Date'], RV['Ozone']):
    plt.text(x = x, y = y, s = '{:.3f}'.format(y))
plt.xticks(RV['Date'],rotation=45, fontsize=15)
plt.show()
```
