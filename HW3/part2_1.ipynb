```python
import pandas as pd
import plotly.figure_factory as ff
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Load in data
data = pd.read_csv('cal_housing.data', header = None)
# Set column names
data.columns = data.columns = ['Longitude','Latitude', 'Households', 'Population', 'Total Bedrooms', 'Total Rooms','Housing Median Age','Median Income','Median House Value']
display(data)
```

```python
# Plot the distribution of median house value.
sns.histplot(data = data, x = 'Median House Value' )
plt.title('Distribution of Median House Value')
```

```python
#Plot median house value vs total bedrooms.
sns.lineplot(x = data["Median House Value"], y = data["Total Bedrooms"])
plt.show()
```

```python
#Plot the correlation matrix.
ax = sns.heatmap(data.corr(), annot=True)
sns.heatmap() # Plot data 3
```
